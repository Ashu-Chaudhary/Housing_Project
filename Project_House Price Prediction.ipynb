```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
sns.distplot(df.SalePrice)
```

```python
df['LogOfPrice'] = np.log(df.SalePrice)
df.drop(["SalePrice"], axis=1, inplace=True)
```

```python
df.skew().sort_values()
```

```python
y = df.LogOfPrice
```

```python
df = df.select_dtypes(include=["int64","float64"]) 
X = df.drop(["LogOfPrice"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
                                    X, y, random_state=3, test_size=.25)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

```python
df = DecisionTreeRegressor()
```

```python
df.fit(X_train,y_train)
```

```python
df.score(X_test,y_test)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.score(X_test,y_test)
```

```python
gd=GradientBoostingRegressor()
```

```python
gd.fit(X_train,y_train)
```

```python
gd.score(X_test,y_test)
```

```python
ad=AdaBoostRegressor()
```

```python
ad.fit(X_train,y_train)
```

```python
ad.score(X_test,y_test)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
ET=ExtraTreesRegressor()
```

```python
ET.fit(X_train,y_train)
```

```python
ET.score(X_test,y_test)
```

```python
print ("R^2 is: \n", gd.score(X_test, y_test))
```

```python
predictions = gd.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print ('RMSE is: \n', mean_squared_error(y_test, predictions))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
gd_cv = cross_val_score(gd, X, y, cv = 10, scoring= 'r2')
print("Cross-validation results: ", gd_cv)
print("R2: ", gd_cv.mean())
```

```python
from sklearn.linear_model import Ridge, Lasso
```
